## Анализ датасета
Загружаем датасет и смотрим, какие есть данные. 
Далее делим датасет на признаки X и таргет y для предсказания y - RainTomorrow

```python
df = df[df.RainTomorrow.notna()]
```

В таргет перенной удалим все отсутствующие значения, так как сравнивать результат иначе будет не с чем

```python
X = df.loc[:, df.columns != 'RainTomorrow']
y = df['RainTomorrow']
X.shape
```

```python
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.25, shuffle=False)
```

Делим выборку на train и test
Далее будем обрабатывать данные, тк из результата describe() видно что есть много неизвестных значений, а также некоторые столбцы состоят из текстовых данных

```python
month_train = pd.to_datetime(X_train.Date).dt.month
month_test = pd.to_datetime(X_test.Date).dt.month
```

```python
# train
location_train = pd.get_dummies(data = X_train['Location'], prefix = 'Location_')
wind_gust_dir_train = pd.get_dummies(data = X_train['WindGustDir'], prefix = 'WindGustDir_')
wind_dir_9am_train = pd.get_dummies(data = X_train['WindDir9am'], prefix = 'WindDir9am_')
wind_dir_3pm_train = pd.get_dummies(data = X_train['WindDir3pm'], prefix = 'WindDir3pm_')

# test
location_test = pd.get_dummies(data = X_test['Location'], prefix = 'Location_')
wind_gust_dir_test = pd.get_dummies(data = X_test['WindGustDir'], prefix = 'WindGustDir_')
wind_dir_9am_test = pd.get_dummies(data = X_test['WindDir9am'], prefix = 'WindDir9am_')
wind_dir_3pm_test = pd.get_dummies(data = X_test['WindDir3pm'], prefix = 'WindDir3pm_')
```

```python
# train
X_train.drop('Date', 1, inplace=True)
X_train.drop('Location', 1, inplace=True)
X_train.drop('WindGustDir', 1, inplace=True)
X_train.drop('WindDir9am', 1, inplace=True)
X_train.drop('WindDir3pm', 1, inplace=True)

# test
X_test.drop('Date', 1, inplace=True)
X_test.drop('Location', 1, inplace=True)
X_test.drop('WindGustDir', 1, inplace=True)
X_test.drop('WindDir9am', 1, inplace=True)
X_test.drop('WindDir3pm', 1, inplace=True)
```

Видим что столбцы RainToday и RainTomorow можно привести к виду бинарных векторов. Заменим отсутствующие значения на средние.

```python
# train
X_train.RainToday[X_train.RainToday == 'No'] = 0
X_train.RainToday[X_train.RainToday == 'Yes'] = 1
X_train.RainToday[X_train.RainToday.isna()] = 1 if (np.mean(X_train.RainToday) >= 0.5) else 0 
y_train[y_train == 'No'] = 0
y_train[y_train == 'Yes'] = 1
X_train.RainToday = X_train.RainToday.astype('int')
y_train=y_train.astype('int')

# test
X_test.RainToday[X_test.RainToday == 'No'] = 0
X_test.RainToday[X_test.RainToday == 'Yes'] = 1
X_test.RainToday[X_test.RainToday.isna()] = 1 if (np.mean(X_train.RainToday) >= 0.5) else 0
y_test[y_test == 'No'] = 0
y_test[y_test == 'Yes'] = 1
X_test.RainToday = X_test.RainToday.astype('int')
y_test=y_test.astype('int')
```

```python
meanMinTemp = X_train['MinTemp'].mean() 
meanMaxTemp = X_train['MaxTemp'].mean() 
meanRainfall = X_train['Rainfall'].mean() 
meanEvaporation = X_train['Evaporation'].mean()  
meanSunshine = X_train['Sunshine'].mean() 
meanPressure9am = X_train['Pressure9am'].mean() 
meanPressure3pm = X_train['Pressure3pm'].mean() 
meanWindSpeed9am = X_train['WindSpeed9am'].mean() 
meanWindGustSpeed = X_train['WindGustSpeed'].mean() 
meanWindSpeed3pm = X_train['WindSpeed3pm'].mean() 
meanHumidity9am = X_train['Humidity9am'].mean() 
meanHumidity3pm = X_train['Humidity3pm'].mean() 
meanCloud9am = X_train['Cloud9am'].mean() 
meanCloud3pm = X_train['Cloud3pm'].mean() 
meanTemp9am = X_train['Temp9am'].mean() 
meanTemp3pm = X_train['Temp3pm'].mean() 
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

X_train['MinTemp']= imp.fit_transform(X_train['MinTemp'].values.reshape(-1, 1))
X_train['MaxTemp']= imp.fit_transform(X_train['MaxTemp'].values.reshape(-1, 1))
X_train['Rainfall']= imp.fit_transform(X_train['Rainfall'].values.reshape(-1, 1))
X_train['Evaporation'] = imp.fit_transform(X_train['Evaporation'].values.reshape(-1, 1))
X_train['Sunshine']= imp.fit_transform(X_train['Sunshine'].values.reshape(-1, 1))
X_train['Pressure9am']= imp.fit_transform(X_train['Pressure9am'].values.reshape(-1, 1))
X_train['Pressure3pm']= imp.fit_transform(X_train['Pressure3pm'].values.reshape(-1, 1))
X_train['WindSpeed9am']= imp.fit_transform(X_train['WindSpeed9am'].values.reshape(-1, 1))
X_train['WindGustSpeed']= imp.fit_transform(X_train['WindGustSpeed'].values.reshape(-1, 1))
X_train['WindSpeed3pm']= imp.fit_transform(X_train['WindSpeed3pm'].values.reshape(-1, 1))
X_train['Humidity9am']= imp.fit_transform(X_train['Humidity9am'].values.reshape(-1, 1))
X_train['Humidity3pm']= imp.fit_transform(X_train['Humidity3pm'].values.reshape(-1, 1))
X_train['Cloud9am']= imp.fit_transform(X_train['Cloud9am'].values.reshape(-1, 1))
X_train['Cloud3pm']= imp.fit_transform(X_train['Cloud3pm'].values.reshape(-1, 1))
X_train['Temp9am']= imp.fit_transform(X_train['Temp9am'].values.reshape(-1, 1))
X_train['Temp3pm']= imp.fit_transform(X_train['Temp3pm'].values.reshape(-1, 1))
```

```python
X_test.MinTemp[X_test.MinTemp.isna()] = meanMinTemp
X_test.MaxTemp[X_test.MaxTemp.isna()] = meanMaxTemp
X_test.Rainfall[X_test.Rainfall.isna()] = meanRainfall
X_test.Evaporation[X_test.Evaporation.isna()] = meanEvaporation
X_test.Sunshine[X_test.Sunshine.isna()] = meanSunshine
X_test.Pressure9am[X_test.Pressure9am.isna()] = meanPressure9am
X_test.Pressure3pm[X_test.Pressure3pm.isna()] = meanPressure3pm
X_test.WindSpeed9am[X_test.WindSpeed9am.isna()] = meanWindSpeed9am
X_test.WindGustSpeed[X_test.WindGustSpeed.isna()] = meanWindGustSpeed
X_test.WindSpeed3pm[X_test.WindSpeed3pm.isna()] = meanWindSpeed3pm
X_test.Humidity9am[X_test.Humidity9am.isna()] = meanHumidity9am
X_test.Humidity3pm[X_test.Humidity3pm.isna()] = meanHumidity3pm
X_test.Pressure9am[X_test.Pressure9am.isna()] = meanPressure9am
X_test.Pressure3pm[X_test.Pressure3pm.isna()] = meanPressure3pm
X_test.Cloud9am[X_test.Cloud9am.isna()] = meanCloud9am
X_test.Cloud3pm[X_test.Cloud3pm.isna()] = meanCloud3pm
X_test.Temp9am[X_test.Temp9am.isna()] = meanTemp9am
X_test.Temp3pm[X_test.Temp3pm.isna()] = meanTemp3pm
```

```python
def logfunc(df, indices):
  res = df.copy()
  for i in indices:
    res.iloc[:,i] = np.log1p(res.iloc[:,i])
  return res
```

```python
%%time
idxs = [1, 2, 3, 4, 5]
from sklearn.naive_bayes import  GaussianNB

classifier = GaussianNB()
classifier.fit(X_train, y_train)
prediction = classifier.predict(X_test)

print (np.mean(prediction == y_test))
```

```python
%%time

classifier = BernoulliNB()
classifier.fit(logfunc(X_train, idxs), y_train)
prediction = classifier.predict(logfunc(X_test, idxs))

print (np.mean(prediction == y_test))
```

```python
%%time

classifier = GaussianNB()
classifier.fit(logfunc(X_train, idxs), y_train)
prediction = classifier.predict(logfunc(X_test, idxs))

print (np.mean(prediction == y_test))
```
